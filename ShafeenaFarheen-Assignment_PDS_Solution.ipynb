# 1.Merge the dataset athletes and regions. 

```python
#Reading the "athlete_events.csv" data set using .read_csv() pandas method and storing the same into "Data_athlete" dataframe
Data_athlete = pd.read_csv("athlete_events.csv")
Data_athlete
```

```python
##Reading the "regions.csv" data set using .read_csv() pandas method and storing the same into "regions" dataframe
regions = pd.read_csv("regions.csv")
regions
```

```python
#The merge () method updates the content of two DataFrame by merging them together, using the specified method (s). 
merged = pd.merge(Data_athlete, regions, on='NOC', how='left')
merged.head()
```

# 5.	Display the top 5 countries who won the most medals. Support your answer with suitable plot. 


```python
##This will count the number of times each value occurs within the column and return the data with the features reset_index(name='Medal').head()
goldMedallist.region.value_counts().reset_index(name='Medal').head()
```

```python
#Displaying the catplot to show the top 5 countries who won the most medals
totalGoldMedals = goldMedallist.region.value_counts().reset_index(name='Medal').head(5)
gold = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
                height=6, kind="bar", palette="prism")
gold.despine(left=True)
gold.set_xlabels("Top 5 countries")
gold.set_ylabels("Number of Medals")
plt.title('Medals per Country')
```

#### Inferences:-The USA seems to be the most winning countryLet’s now review the top 5 gold medal countries:We Displayig the top 5 by using head()
